```python
from tpot import TPOTClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# Import data

```python
# NOTE: Make sure that the outcome column is labeled 'target' in the data file
tpot_data = pd.read_csv('00_datasets/breast_cancer_classification/wdbc_data.csv', 
                        sep=',')
tpot_data.head()
```

## Rename target column

```python
tpot_data = tpot_data.drop('id', axis=1)
```

```python
tpot_data.rename(columns={'diagnosis':'target'}, inplace=True)
tpot_data.head()
```

## Extract features

```python
features = tpot_data.drop('target', axis=1)
features = features.astype(np.float64) # to avoid warning message during optimization
features.head()
```

```python
training_features, testing_features, training_target, testing_target = \
            train_test_split(features, tpot_data['target'], random_state=42)
```

# Perform AutoML with TPOT
See the documentation for complete set of parameters:  
https://epistasislab.github.io/tpot/api/

```python
pipeline_optimizer = TPOTClassifier(generations=5,
                                    population_size=20,
                                    cv=5,                  # cross validation
                                    random_state=42,
                                    verbosity=2)
```

```python
pipeline_optimizer.fit(features=training_features, target=training_target)
```
